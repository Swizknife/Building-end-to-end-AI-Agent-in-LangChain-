```python
#for chat converstaion 
from langchain_ollama import ChatOllama

#intialize the model
llm=ChatOllama(
    model="llama3.2:1b",
    temperature=0,
    #other_params...
)
#simple invocation
response = llm.invoke("explain the concept of qunatum entanglement in one sentence.")
print(response.content)
```

```python
#for non chat execution
from langchain_ollama import OllamaLLM
llm = OllamaLLM(model="llama3.2:1b")
response = llm.invoke("the capital of france is")
print(response)
```

```python
from langchain_ollama import OllamaEmbeddings
embeddings= OllamaEmbeddings(
    model="embeddinggemma:latest",
)
#embed a single query
query_result = embeddings.embed_query("what is langchain?")
print(query_result)

```

```python
#embed multiple documents 
doc_results = embeddings.embed_documents(["Document 1 content", "Document 2 content'])
print(f"Embedding length: {len(doc_results)}")
```
